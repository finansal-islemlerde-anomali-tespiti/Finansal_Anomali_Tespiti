```python
import pandas as pd
import numpy as np

# Temizlenmiş veriyi yüklüyoruz
df = pd.read_csv('../data_generation/output/transactions.csv')
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
print(df.columns)
```

```python
# Sadece Helin'in sistemde 'Tutar Anomalisi' olarak etiketlediği satırları filtreliyoruz
tutar_anomalileri = df[df['anomaly_type'] == 'Tutar Anomalisi']

print(f"=== TESPİT EDİLEN TUTAR ANOMALİLERİ ===")
print(f"Toplam Yakalanan Şüpheli İşlem: {len(tutar_anomalileri)} adet\n")

# En yüksek tutarlı ilk 10 şüpheli işlemi listele
display(tutar_anomalileri[['transaction_id', 'card_id', 'amount','atm_id', 'transaction_type', 'status', 'created_at']].sort_values(by='amount', ascending=False).head(10))
```
