```python
import pandas as pd
import numpy as np

# 1. Hücredeki dosya yolunu temizlenmiş veriyle değiştiriyoruz
df = pd.read_csv('../data/transactions_cleaned.csv') 
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
print(df.columns)
```

```python
# Ham veride 'anomaly_type' olmadığı için tutar limitine göre kendimiz filtreliyoruz
# Örneğin: Ortalama işlem tutarının 5 katını aşan durumlar (Tutar Anomalisi kuralları gibi)
ortalama_tutar = df['amount'].mean()
tutar_anomalileri = df[df['amount'] > (ortalama_tutar * 5)]

print("--- TESPİT EDİLEN TUTAR ANOMALİLERİ ---")
print(f"Toplam Yakalanan Şüpheli İşlem: {len(tutar_anomalileri)} adet\n")

# En yüksek tutarlı ilk 10 şüpheli işlemi listele
display(tutar_anomalileri[['transaction_id', 'card_id', 'atm_id', 'amount', 'transaction_type', 'status', 'created_at']].sort_values(by='amount', ascending=False).head(10))
```
