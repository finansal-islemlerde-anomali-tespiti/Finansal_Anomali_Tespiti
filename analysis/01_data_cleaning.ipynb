```python
import pandas as pd
import numpy as np

# 1. Üretilen 50.400 satırlık ham veriyi yükleyelim
df_tx = pd.read_csv("../data_generation/output/transactions.csv")

# 2. İlk 5 satırı kontrol edelim
print(f"📊 Veri Boyutu: {df_tx.shape}")
display(df_tx.head())

# 3. Veri tiplerini ve genel sağlık durumunu inceleyelim
print("\n--- Sütun Bilgileri ve Tipler ---")
df_tx.info()
```

```python
# 1. created_at sütununu metinden gerçek datetime objesine dönüştürelim
df_tx['created_at'] = pd.to_datetime(df_tx['created_at'])

# 2. İşlemleri en eski tarihten en yeniye doğru kronolojik olarak sıralayalım
# (Bu sıralama pencere fonksiyonlu SQL anomali sorguları için çok önemlidir)
df_tx = df_tx.sort_values(by='created_at').reset_index(drop=True)

# 3. Yaptığımız değişikliği tip kontrolüyle doğrulayalım
print("📅 Güncel Tarih Sınırları:")
print(f"En Eski İşlem: {df_tx['created_at'].min()}")
print(f"En Yeni İşlem: {df_tx['created_at'].max()}")
print("\n--- Yeni Sütun Tipleri ---")
df_tx.info()
```

```python
# 1. Tutar sütununu virgülden sonra tam 2 basamak olacak şekilde kesinleştirelim
df_tx['amount'] = df_tx['amount'].round(2)

# 2. Genel bir istatistik özeti çıkaralım (Burada en yüksek harcama anomalilerimiz de kendini belli edecek)
print("💰 İşlem Tutarları İstatistiki Özeti:")
display(df_tx['amount'].describe())

# 3. Temizlenmiş ve sıralanmış bu veriyi yeni bir CSV dosyası olarak kaydedelim
# Gurbet bu dosyayı doğrudan SQL Server'a import edebilecek
df_tx.to_csv("../data_generation/output/transactions_cleaned.csv", index=False)

print("\n🎉 Tebrikler! Temizlenmiş veri 'transactions_cleaned.csv' adıyla başarıyla kaydedildi.")
```
